# I. Numpy

### Импортируйте NumPy

```python
import numpy as np

def print2(*args):
    for arg in args:
        print(arg)
    print('<-------------------------->')
```

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
arr = np.zeros(10)
arr[(len(arr) - 1) // 2] = 1
print2(arr)

arr = arr.reshape(2,5)
print2(arr)
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
arr = np.linspace(10, 49, 40)
print2(arr)

arr = arr[::-1]
print2(arr)

even_arr = arr[arr % 2 == 0]
print2(even_arr)
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
arr = np.arange(9).reshape(3, 3)
print2(arr)
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
arr = np.random.random((4,3,2))
print2(arr)

min_val = np.min(arr)
max_val = np.max(arr)
print2(min_val, max_val)
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
arr = np.random.randint(1, 10, size=(6,4))
arr_two = np.random.randint(1, 10, size=(4,3))

# можно с оператором @
matrix_multiply = np.dot(arr, arr_two)
print2(matrix_multiply)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
np.random.seed(42)
arr = np.random.random((7,7))
print2(arr)

mean_arr = np.mean(arr)
std_arr = np.std(arr)
print2("Среднее", mean_arr, "Стандартное", std_arr)

min_val = np.min(arr)
max_val = np.max(arr)
normalized_arr = (arr - min_val) / (max_val - min_val)
print2("Нормализация (min-max)", normalized_arr)

normalized_arr2 = (arr - mean_arr) / std_arr
print2("Нормализация (поиск выбросов)", normalized_arr)
```

# II. Pandas

### Импортируйте: pandas, matplotlib, seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)
sns.set_theme()
```

### Загрузите датасет Tips из набора датасетов seaborn

```python
tips = sns.load_dataset("tips")
```

### Посмотрите на первые 5 строчек

```python
tips.head()
```

### Узнайте сколько всего строчек и колонок в данных

```python
print2("Строчки", len(tips))
print2("Колонки", len(tips.columns))
```

### Проверьте есть ли пропуски в данных

```python
print2(tips.isna().sum())
```

### Посмотрите на распределение числовых признаков

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

tips['total_bill'].hist(ax=axes[0], bins=10, edgecolor='black')
axes[0].set_title('Распределение total_bill')
axes[0].set_xlabel('total_bill')
axes[0].set_ylabel('Частота')

tips['tip'].hist(ax=axes[1], bins=10, edgecolor='black', color='orange')
axes[1].set_title('Распределение tip')
axes[1].set_xlabel('tip')
axes[1].set_ylabel('Частота')

tips['size'].hist(ax=axes[2], bins=5, edgecolor='black', color='green')
axes[2].set_title('Распределение size')
axes[2].set_xlabel('size')
axes[2].set_ylabel('Частота')

plt.tight_layout()
plt.show()
```

### Найдите максимальное значение 'total_bill'

```python
print2(tips['total_bill'].max())
```

### Найдите количество курящих людей

```python
print2(tips[tips['smoker'] == 'Yes']['smoker'].count())
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
print2(tips.groupby('day', observed=True)['total_bill'].mean())
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
print2(tips[tips['total_bill'] > tips['total_bill'].median()].groupby('sex', observed=True)['tip'].mean())
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
tips['smoker'] = np.where(tips['smoker'] == 'Yes', 1, 0)
tips.head()
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
plt.figure(figsize=(12, 8))
tips['total_bill'].plot(kind='hist', bins=5, alpha=0.7, color='yellow', edgecolor='black')
plt.title("Распределение значений 'total_bill'")
plt.xlabel('Сумма')
plt.ylabel("Количество")
plt.grid(True)
plt.show()
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
tips.plot.scatter(x='total_bill', y='tip', c=np.random.uniform(15, 80, len(tips)), title='scatterplot между total_bill и tip', figsize=(12, 8))
```

### Постройте pairplot

```python
sns.pairplot(tips)
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
plt.figure(figsize=(8, 5))
tips.boxplot(column='total_bill', by='day', grid=False)
plt.title("Распределение total_bill по дням недели")
plt.suptitle("")
plt.xlabel("day")
plt.ylabel("total_bill")
plt.show()

avg_total_bill = tips.groupby("day")["total_bill"].mean()

plt.figure(figsize=(8, 5))
avg_total_bill.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title("Средний total_bill по дням недели")
plt.xlabel("day")
plt.ylabel("Средний total_bill")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
sns.displot(tips, x='tip', hue='time')
sns.displot(tips, x='tip', hue='time', stat='frequency')
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
sns.relplot(data=tips, x="total_bill", y="tip", col="sex", hue="smoker", kind="scatter")
```
